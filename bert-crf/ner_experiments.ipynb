```python
import torch

from re_utils.train import train_ner
```

```python
num_labels = 17
train_tokenized_texts_path = "resources/data/train/labeled_texts.jsonl"
test_tokenized_texts_path = "resources/data/test/labeled_texts.jsonl"
dropout = 0.2
batch_size = 16
epochs = 15
log_every = 20
lr_new_layers = 1e-3
lr_bert = 1e-5
device = "cpu"
# device = "mps" if torch.backends.mps.is_available() else "cpu"
```

```python
# sberbank-ai/ruBert-base w/o crf

train_ner(
    num_labels=num_labels,
    bert_name="sberbank-ai/ruBert-base",
    train_tokenized_texts_path=train_tokenized_texts_path,
    test_tokenized_texts_path=test_tokenized_texts_path,
    dropout=dropout,
    batch_size=batch_size,
    epochs=epochs,
    log_every=log_every,
    lr_new_layers=lr_new_layers,
    lr_bert=lr_bert,
    use_crf=False,
    save_to="weights/bert-without-crf.pt",
    device=device
)
```

```python
# bert-base-uncased+ crf

train_ner(
    num_labels=63,
    bert_name="bert-base-uncased",
    train_tokenized_texts_path="train_processed.jsonl",
    test_tokenized_texts_path="test_processed.jsonl",
    dropout=dropout,
    batch_size=batch_size,
    epochs=epochs,
    log_every=log_every,
    lr_new_layers=lr_new_layers,
    lr_bert=lr_bert,
    use_crf=True,
    save_to="weights/bert-with-crf-eng-0414-01.pt",
    device=device
)
```

```python
# sberbank-ai/ruBert-base + crf

train_ner(
    num_labels=num_labels,
    bert_name="sberbank-ai/ruBert-base",
    train_tokenized_texts_path=train_tokenized_texts_path,
    test_tokenized_texts_path=test_tokenized_texts_path,
    dropout=dropout,
    batch_size=batch_size,
    epochs=epochs,
    log_every=log_every,
    lr_new_layers=lr_new_layers,
    lr_bert=lr_bert,
    use_crf=True,
    save_to="weights/bert-with-crf.pt",
    device=device
)
```

```python
# rurebus-bert w/o crf

train_ner(
    num_labels=num_labels,
    bert_name="weights/rurebus-bert",
    train_tokenized_texts_path=train_tokenized_texts_path,
    test_tokenized_texts_path=test_tokenized_texts_path,
    dropout=dropout,
    batch_size=batch_size,
    epochs=epochs,
    log_every=log_every,
    lr_new_layers=lr_new_layers,
    lr_bert=lr_bert,
    use_crf=False,
    save_to="weights/bert-crf.pt",
    device=device
)
```

```python
# rurebus-bert + crf

train_ner(
    num_labels=num_labels,
    bert_name="weights/rurebus-bert",
    train_tokenized_texts_path=train_tokenized_texts_path,
    test_tokenized_texts_path=test_tokenized_texts_path,
    dropout=dropout,
    batch_size=batch_size,
    epochs=epochs,
    log_every=log_every,
    lr_new_layers=lr_new_layers,
    lr_bert=lr_bert,
    use_crf=True,
    save_to="weights/bert-wo-crf.pt",
    device=device
)
```
